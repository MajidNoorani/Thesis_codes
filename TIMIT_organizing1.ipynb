```python
#  In this code we read a part of training files and split them into two categories
#  Females and Males
#  Then we save them in specified folders in continues enumerating
```

```python
import os
import soundfile as sf
import time
import math
import numpy as np
```

```python
path0 = [x[0] for x in os.walk('TIMIT/TRAIN')]
path1 = [x[1] for x in os.walk('TIMIT/TRAIN')]
files = [x[2] for x in os.walk('TIMIT/TRAIN')]
```

```python
path0
```

```python
# extracting addresses of male and female speakers and their sentences
DR8_male_speakers = []
for addr in path0:
    if 'DR8/M' in addr:
        DR8_male_speakers.append(addr)
        
DR8_female_speakers = []
for addr in path0:
    if 'DR8/F' in addr:
        DR8_female_speakers.append(addr)
```

```python
print(len(DR8_male_speakers))
print(len(DR8_female_speakers))
```

```python
# Extracting train and test data of male speakers and saving them in folders 'Male' nad 'Male_test' 
sentences = []
counter = 0
for person in DR8_male_speakers:
    files = [x[2] for x in os.walk(person)]
    for file in np.squeeze(files):
        if '.WAV' in file:
            sentences.append(person + '/' + file)
    for i in range(8):
        data, samplerate = sf.read(sentences[i])
        sf.write('TIMIT/Organized/Male/' +'M' + str(i + 8*counter) + '.WAV', data, samplerate)
    for i in range(8,10):
        data, samplerate = sf.read(sentences[i])
        sf.write('TIMIT/Organized/Male_test/' +'M' + str(i + 2*counter) + '.WAV', data, samplerate)
    counter = counter + 1
    sentences = []
        
    
```

```python
# Extracting train and test data of female speakers and saving them in folders 'Female' nad 'Female_test' 
sentences = []
counter = 0
for person in DR8_female_speakers:
    files = [x[2] for x in os.walk(person)]
    for file in np.squeeze(files):
        if '.WAV' in file:
            sentences.append(person + '/' + file)
    for i in range(8):
        data, samplerate = sf.read(sentences[i])
        sf.write('TIMIT/Organized/Female/' +'F' + str(i + 8*counter) + '.WAV', data, samplerate)
    for i in range(8,10):
        data, samplerate = sf.read(sentences[i])
        sf.write('TIMIT/Organized/Female_test/' +'F' + str(i + 2*counter) + '.WAV', data, samplerate)
    counter = counter + 1
    sentences = []
        
    
```

```python
male_sentences = []
for male in DR8_male_speakers:
    sentences = [x[2] for x in os.walk(male)]
    for sentence in np.squeeze(sentences):
        if '.WAV' in sentence:
            male_sentences.append(male + '/' + sentence)
            
female_sentences = []
for female in DR8_female_speakers:
    sentences = [x[2] for x in os.walk(female)]
    for sentence in np.squeeze(sentences):
        if '.WAV' in sentence:
            female_sentences.append(female + '/' + sentence)
```

```python
print(len(male_sentences))
print(len(female_sentences))
```

```python
for i in range(len(male_sentences)):
    if i%8!=0 and i%9!=0:
        data, samplerate = sf.read(male_sentences[i])
        sf.write('TIMIT/Organized/Female/' +'F' + str(i) + '.WAV', data, samplerate)
        i += 1
    
```
