```python
data_series = str(4)
```

```python
h5f = h5py.File('TIMIT/Organized/concatenated/mixed'  + data_series + '_shuffled.hdf5','r')
X = h5f['mixed'  + data_series + '_shuffled'][:]
h5f.close()
```

```python
h5f = h5py.File('TIMIT/Organized/concatenated/clean'  + data_series + '_shuffled.hdf5','r')
y = h5f['clean'  + data_series + '_shuffled'][:]
h5f.close()
```

```python
X_dim= 257
y_dim = 514
def DNN_A():
    # create model
    model = Sequential()
    model.add(Dense(h_A[0], input_dim = X_dim, kernel_initializer='normal', activation='relu'))
    model.add(Dense(h_A[1], kernel_initializer='normal', activation='relu'))
    model.add(Dense(h_A[2], kernel_initializer='normal', activation='relu'))
    model.add(Dense(y_dim, kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

```python
# seed = 1
# np.random.seed(seed)
print("^^^^^^^^^^^^^^^^^^^^^ Training with 5-fold on mixed0^^^^^^^^^^^^^^^^^^^^^")
tic = time.time()
estimator_A = KerasRegressor(build_fn=DNN_A, epochs=50, batch_size=100, verbose=1)
# kfold = KFold(n_splits=2, random_state=seed)
kfold = KFold(n_splits=5)
results = cross_val_score(estimator_A, X.T, y.T, cv=kfold)
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
toc_fold = time.time()

print("^^^^^^^^^^^^^^^^^^^^^ Training on mixed0^^^^^^^^^^^^^^^^^^^^^")
estimator_A.fit(X.T, y.T)
# prediction = estimator_A.predict(X_test)

toc_fit = time.time()
```

```python
print("tIME OF K-FOLD=", (toc_fold - tic)/60)
print("tIME OF FIT=", (toc_fit - tic)/60)
```
