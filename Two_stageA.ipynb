```python
def reconstruct(wave,angle):
    recon1 = wave*np.cos(angle)+wave*np.sin(angle)*1j
#     recon = np.sqrt(np.power(10, wave))
#     recon1 = recon*np.cos(angle)+recon*np.sin(angle)*1j
    recon = librosa.core.istft((recon1.T), hop_length=160, win_length=400, window='hann')
    return recon
```

```python
def da(load_size, data_len):
    global I
    h5f1 = h5py.File('TIMIT/Organized/concatenated/mixed'  + data_series + '.hdf5','r')
    h5f2 = h5py.File('TIMIT/Organized/concatenated/clean'  + data_series + '.hdf5','r')
    indx = np.arange(0, data_len, load_size)
    indx = shuffle(indx, random_state=0)
    print(indx)
    while True:             #this line is just because keras needs infinite generators
        for I in indx: 
            if data_len-I<load_size:
                Data1 = h5f1['mixed' + data_series][:,I:].T
                Data2 = h5f2['clean' + data_series][:,I:].T
            else:
                Data1 = h5f1['mixed' + data_series][:,I:I+load_size].T
                Data2 = h5f2['clean' + data_series][:,I:I+load_size].T
            a = Data2[:,0:257]
            b = Data2[:,257:]
            Mask = a/Data1
            Mask = np.concatenate((Mask,b/Data1), axis = 1)
#             Mask = sklearn.preprocessing.normalize(Mask, norm='l2', axis=1, copy=True, return_norm=False)
#             Data1 = sklearn.preprocessing.normalize(Data1, norm='l2', axis=1, copy=True, return_norm=False)
            Data1 = (Data1-np.mean(Data1))/(np.max(Data1) - np.min(Data1))
            Mask = (Mask-np.mean(Mask))/(np.max(Mask) - np.min(Mask))
            yield(Data1, Mask)
```

```python
X_dim= 1025
y_dim = 2050
h_A = [1025, 1025, 1025]
def DNN_A():
    # create model
    model = Sequential()
    model.add(Dense(h_A[0], input_dim = X_dim, kernel_initializer='glorot_normal'))
#     keras.layers.Dropout(0.8, noise_shape=None, seed=None)
    model.add(BatchNormalization())
    model.add(Activation('sigmoid'))
#     model.add(LeakyReLU(alpha=0.1))

    model.add(Dense(h_A[1], kernel_initializer='glorot_normal'))
#     keras.layers.Dropout(0.8, noise_shape=None, seed=None)
    model.add(BatchNormalization())
    model.add(Activation('sigmoid'))
#     model.add(LeakyReLU(alpha=0.1))

    model.add(Dense(h_A[2], kernel_initializer='glorot_normal'))
    keras.layers.Dropout(0.8, noise_shape=None, seed=None)
    model.add(BatchNormalization())
    model.add(Activation('sigmoid'))
#     model.add(LeakyReLU(alpha=0.1))
    
    model.add(Dense(y_dim, kernel_initializer='glorot_normal'))
#     keras.layers.Dropout(0.8, noise_shape=None, seed=None)
    model.add(BatchNormalization())
    model.add(Activation('sigmoid'))
#     model.add(LeakyReLU(alpha=0.1))

    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```
